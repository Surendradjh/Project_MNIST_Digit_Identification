```python
from keras.datasets import mnist
```

```python
(x_train,y_train),(x_test,y_test)=mnist.load_data()
```

```python
x_train.shape
```

```python
from keras.models import Sequential
from keras.layers import InputLayer,Dense,Flatten,Conv2D,MaxPooling2D
```

```python
model=Sequential()
model.add(InputLayer(shape=(28,28,1)))
model.add(Conv2D(28,(3,3),strides=(1,1),activation='relu',padding='valid',kernel_initializer='he_normal'))
model.add(MaxPooling2D((2,2),padding='valid'))
model.add(Conv2D(28,(3,3),strides=(1,1),activation='relu',padding='valid',kernel_initializer='he_normal'))
model.add(MaxPooling2D((2,2),padding='valid'))
model.add(Conv2D(28,(3,3),strides=(1,1),activation='relu',padding='valid',kernel_initializer='he_normal'))
model.add(MaxPooling2D((2,2),padding='valid'))
model.add(Flatten())
model.add(Dense(units=128,activation='relu'))
model.add(Dense(units=65,activation='relu'))
model.add(Dense(units=30,activation='relu'))
model.add(Dense(units=10,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint

checkpoint = ModelCheckpoint(
    filepath="/content/drive/MyDrive/DL/mnist_model.keras",
    monitor='val_accuracy',
    save_best_only=True,
    save_weights_only=False,
    mode='max',
    verbose=1
)
```

```python
model.fit(x_train,y_train,epochs=20,batch_size=100,validation_data=(x_test,y_test),callbacks=[checkpoint])
```
